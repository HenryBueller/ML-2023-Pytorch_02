### 随机抽样

````python
# ---------- 随机抽样 random sampling ----------
# from sklearn.model_selection import train_test_split
# train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

# 以下shuffle_and_split_data(data, test_ratio)为该函数原理
# 区别：
# 1. train_test_split()中random_state为随机种子
# 2. train_test_split()能传递多个有相同行数的数据集，会拆分到相同索引

# import numpy as np
# def shuffle_and_split_data(data, test_ratio):
#     shuffled_indices = np.random.permutation(len(data))
#     test_set_size = int(len(data) * test_ratio)
#     test_indices = shuffled_indices[:test_set_size]
#     train_indices = shuffled_indices[test_set_size:]
#     return data.iloc[train_indices], data.iloc[test_indices]

# You can then use this function like this:
# >>> train_set, test_set = shuffle_and_split_data(housing, 0.2)
# >>> len(train_set)
# 16512
# >>> len(test_set)
# 4128

# 每次数据集都不一样，解决：保存数据、设置随机种子np.random.seed(42)
# 但是数据集更新之后无效，解决：唯一标识值+哈希，取小于等于最大哈希值的20%的哈希

import numpy as np
from zlib import crc32
def is_id_in_test_set(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32 # python3.0后为无符号数，故&
def split_data_with_id_hash(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: is_id_in_test_set(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]

# 无标识的需要序号（注意增删后序号变化影响）、某一特殊变量、某几个特殊变量组合作为唯一标识
# # 【法1】序号
# housing_with_id = housing.reset_index() # adds an `index` column
# train_set, test_set = split_data_with_id_hash(housing_with_id, 0.2, "index")
# # 【法2】经纬度
housing_with_id = housing.reset_index()  # copy & adds an `index` column
housing_with_id["id"] = housing["longitude"] * 1000 + housing["latitude"]
train_set, test_set = split_data_with_id_hash(housing_with_id, 0.2, "id")
````

### 分层抽样

````python
# ---------- 分层抽样 stratified sampling ----------

# 假设中等收入对房价影响很重要，阶层数目不能太多（保证每个阶层stratum都有充足样例数量）
# 增加辅助的属性income_cat

housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
housing["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.xlabel("Income category")
plt.ylabel("Number of districts")
plt.show()
````

````python
from sklearn.model_selection import StratifiedShuffleSplit
# 【法1】10个不同的分层分割，取第一个（交叉验证）
splitter = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
strat_splits = []
for train_index, test_index in splitter.split(housing, housing["income_cat"]): # split产生的是序号
    strat_train_set_n = housing.iloc[train_index].copy()
    strat_test_set_n = housing.iloc[test_index].copy()
    strat_splits.append([strat_train_set_n, strat_test_set_n])
strat_train_set, strat_test_set = strat_splits[0]
# 【法2】或者直接只取一个
# strat_train_set, strat_test_set = train_test_split(housing,test_size=0.2,
#                                                    stratify=housing["income_cat"], random_state=42)
````

````python
strat_test_set["income_cat"].value_counts()
````

````python
# extra code – computes the data for Figure 2–10
from sklearn.model_selection import train_test_split

def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(housing),
    "Stratified %": income_cat_proportions(strat_test_set),
    "Random %": income_cat_proportions(test_set),
}).sort_index()
compare_props.index.name = "Income Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)
````

````python
# 删除辅助的属性income_cat，因为.copy()，housing仍有该属性
# print(strat_train_set.head())
# print(strat_test_set.head())
strat_train_set.drop("income_cat", axis=1, inplace=True)
strat_test_set.drop("income_cat", axis=1, inplace=True)
````

````python
# option 3 实际使用：填入中位数

# 使用SimpleImputer算出中位数
# (strategy="median"), or (strategy="mean"), 
# or (strategy="most_frequent"), or (strategy="constant", fill_value=0)
# 后两者支持非数字
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median") # 创建一个SimpleImputer实例
# print(imputer.strategy)
housing_num = housing.select_dtypes(include=[np.number]) # copy with only the numerical attributes
imputer.fit(housing_num) # calculate
# print(imputer.statistics_) # array([-118.51 , 34.26 , 29. , 2125. , 434. , 1167. , 408. , 3.5385])
# print(housing_num.median().values) # array([-118.51 , 34.26 , 29. , 2125. , 434. , 1167. , 408. , 3.5385])

# 处理训练数据集
X = imputer.transform(housing_num) # NumPy arrays
# print(imputer.feature_names_in_) 
# # array(['longitude', 'latitude', 'housing_median_age', 
# #        'total_rooms', 'total_bedrooms', 'population', 
# #        'households', 'median_income'],dtype=object)
housing_tr = pd.DataFrame(X, columns=housing_num.columns, index=housing_num.index) # 转回pd

# There are also more powerful imputers available in the sklearn.impute package (both for numerical features only):
# • KNNImputer replaces each missing value with the mean of the k-nearest neighbors’ values for that feature. 
# The distance is based on all the available features.
# • IterativeImputer trains a regression model per feature to predict the missing values based on all the other available features. 
# It then trains the model again on the updated data, 
# and repeats the process several times, improving the models and the replacement values at each iteration.
````

### 特征缩放和变形

> 特征缩放重中之重，没有很难表现好  
> 只能用训练数据调整scaler，在训练数据`fit()`or`fit_transform()`后在其他数据`transform()`  
> 为防止新数据超范围，`clip=True`  
>  
> 神经网络均值=0效果最好  
> 标准化没有具体范围，因此outlier影响比正则化小  

````python
# ----- Min-max scaling (normalization)：[0-1] (cur_x-min)/(max-min)-----
from sklearn.preprocessing import MinMaxScaler # MinMaxScaler default 0-1
min_max_scaler = MinMaxScaler(feature_range=(-1, 1)) # neural networks work best with zero-mean inputs
housing_num_min_max_scaled = min_max_scaler.fit_transform(housing_num)

# ----- Standardization (cur_x-mean), then standard deviation -----
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
housing_num_std_scaled = std_scaler.fit_transform(housing_num)

# 如果想scale稀疏矩阵，需要先转化成密集矩阵。
# 如果想直接求，使用`StandardScaler(with_mean=False)`就不减平均值直接standard deviation，
# 此时稀疏矩阵不需要转化成密集矩阵（稀疏矩阵存的时候只记录非零信息）
````

<mark>`fit()`和`fit_transform()`只能用于训练集，之后遇到只能`transform()`，以保证一致性</mark>  
当特征分布远离均值的数据不是指数级稀少的时候heavy tail，
正则化和标准化都会把多数数据压到很小范围内，因此在<mark>缩放特征前要先变形</mark>对heavy tail脱水,可能的话图像变成大致对称  
$\qquad$1 正数+heavy tail右：平方根，或者其他0-1的指数  
$\qquad$2 heavy tail很长，如幂律分布：取对数  
$\qquad$$\qquad$桶化bucketizing the feature：扔到大小大致相等的桶里，用桶编号代替（如用每个值百分位数percentile代替值）  
$\qquad$$\qquad$平均分布没必要进一步scale，想的话根据桶的数量把值映射到0-1  
$\qquad$3 多模态分布即多峰分布multimodal distribution：桶化  
$\qquad$$\qquad$把`housing_median_age`值作为类而不是数值桶化，桶序号编码成独热矩阵
